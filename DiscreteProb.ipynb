```r
set.seed(105)
n = 100

probs = c(24/36, 8/36, 4/36)
events <- c('Lose', 'Win', 'Win Big')

sim <- sample(events, n, probs, replace = TRUE)

counts <- table(sim)
counts
barplot(counts,
        main = "Simulated Counts of Events",
        col = "skyblue",
        ylab = "Frequency",
        xlab = "Event")
```

```r
# Assign payouts (example: A = -1, B = +2, C = -3)
payouts <- c(Lose = -10, Win = 10, `Win Big` = 50)

# Map outcomes to payouts
winnings <- payouts[sim]

# Compute running total (cumulative sum)
running_total <- cumsum(winnings)

# Plot running total
plot(running_total, type = "l",
     main = "Running Total of Craps Strategy",
     xlab = "Number of Trials",
     ylab = "Cumulative Winnings",
     col = "blue")
abline(h = 0, lty = 2, col = "red")  # zero reference line
```

```r
simulate_once <- function(n = 1000) {
  sim <- sample(events, size = n, replace = TRUE, prob = probs)
  winnings <- payouts[sim]
  sum(winnings)  # final total after n trials
}

# Run simulation 1,000 times
n_runs <- 1000
final_totals <- replicate(n_runs, simulate_once())

# Histogram of results
hist(final_totals,
     breaks = 50,
     col = "lightblue",
     main = "Distribution of Final Totals",
     xlab = "Final Bankroll after 1000 Trials")
```
